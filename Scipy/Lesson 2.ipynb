```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#Probability Density Function
x = np.linspace(-6,-6,1000)
fx = norm.pdf(x,loc= 0,scale =1)
```

```python
plt.plot(x,fx);
```

```python
#cummulative Density Function
FX = norm.cdf(x,loc=0, scale =1)
plt.plot(x,FX);
```

```python
#CONVOLUTION FOR Blurry filter
from PIL import Image
```

```python
#import local image
filepath = 'D:/GIT1/Machine Learning/ML Libraries/Machine-Learning-Libraries/Matplot/mona.jpeg'
```

```python
#open image
im = Image.open(filepath)
im
```

```python
#convert to greyscale
gray = np.mean(im,axis = 2)
gray
```

```python
#blurry cn
x = np.linspace(-6,6,50)
fx = norm.pdf(x,loc = 0,scale =1)
```

```python
filt = np.outer(fx,x)
plt.imshow(filt,cmap='gray')
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray, filt)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray,cmap ='gray')
plt.subplot(1,2,2)
plt.imshow(out,cmap ='gray')
```
