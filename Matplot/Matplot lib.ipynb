```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Line charts -- generate randomly the x and y axis
x =np.linspace(0,20,1000)
y = np.sin(x)+0.2*x
```

```python
#plotting the axis
plt.plot(x,y)
```

```python
#Axis labels
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('My plot');
```
